```python
# Load the csv FastFoodNutritionMenuV2 from the resources folder. 

fastfooddata = pd.read_csv('../resources/FastFoodNutritionMenuV2.csv')

# display the first 5 rows of the data
fastfooddata.head()
```

```python
# import the data into a sqlite database
conn = sqlite3.connect('fastfood.db')
fastfooddata.to_sql('menu', conn, if_exists='replace', index=False)
# display the first 5 rows of the data from the sqlite database
query = 'SELECT * FROM menu LIMIT 5'
pd.read_sql(query, conn)
```

```python
# create the following queries: 
# 1. Find the average calories for each category
query1 = 'SELECT Category, AVG(Calories) AS AvgCalories FROM menu GROUP BY Category'
# 2. Find the average calories for each restaurant
query2 = 'SELECT Restaurant, AVG(Calories) AS AvgCalories FROM menu GROUP BY Restaurant'
# 3. Find the average calories for each category and restaurant
query3 = 'SELECT Category, Restaurant, AVG(Calories) AS AvgCalories FROM menu GROUP BY Category, Restaurant'
# 4. Find the average calories for each category and restaurant where the calories are greater than 500
query4 = 'SELECT Category, Restaurant, AVG(Calories) AS AvgCalories FROM menu WHERE Calories > 500 GROUP BY Category, Restaurant'
# 5. Find the average calories for each category and restaurant where the calories are greater than 500 and the protein is greater than 20
query5 = 'SELECT Category, Restaurant, AVG(Calories) AS AvgCalories FROM menu WHERE Calories > 500 AND Protein > 20 GROUP BY Category, Restaurant'
# 6. Find the average calories for each category and restaurant where the calories are greater than 500 and the protein is greater than 20 and the carbohydrates are less than 50
query6 = 'SELECT Category, Restaurant, AVG(Calories) AS AvgCalories FROM menu WHERE Calories > 500 AND Protein > 20 AND Carbohydrates < 50 GROUP BY Category, Restaurant'
# 7. Find the average calories for each category and restaurant where the calories are greater than 500 and the protein is greater than 20 and the carbohydrates are less than 50 and the sodium is less than 1000
query7 = 'SELECT Category, Restaurant, AVG(Calories) AS AvgCalories FROM menu WHERE Calories > 500 AND Protein > 20 AND Carbohydrates < 50 AND Sodium < 1000 GROUP BY Category, Restaurant'
# 8. Find the average calories for each category and restaurant where the calories are greater than 500 and the protein is greater than 20 and the carbohydrates are less than 50 and the sodium is less than 1000 and the cholesterol is less than 100
query8 = 'SELECT Category, Restaurant, AVG(Calories) AS AvgCalories FROM menu WHERE Calories > 500 AND Protein > 20 AND Carbohydrates < 50 AND Sodium < 1000 AND Cholesterol < 100 GROUP BY Category, Restaurant'
# 9. Find the average calories for each category and restaurant where the calories are greater than 500 and the protein is greater than 20 and the carbohydrates are less than 50 and the sodium is less than 1000 and the cholesterol is less than 100 and the sugars are less than 50
query9 = 'SELECT Category, Restaurant, AVG(Calories) AS AvgCalories FROM menu WHERE Calories > 500 AND Protein > 20 AND Carbohydrates < 50 AND Sodium < 1000 AND Cholesterol < 100 AND Sugars < 50 GROUP BY Category, Restaurant'
# 10. Find the average calories for each category and restaurant where the calories are greater than 500 and the protein is greater than 20 and the carbohydrates are less than 50 and the sodium is less than 1000 and the cholesterol is less than 100 and the sugars are less than 50 and the fiber is greater than 5
query10 = 'SELECT Category, Restaurant, AVG(Calories) AS AvgCalories FROM menu WHERE Calories > 500 AND Protein > 20 AND Carbohydrates < 50 AND Sodium < 1000 AND Cholesterol < 100 AND Sugars < 50 AND Fiber > 5 GROUP BY Category, Restaurant'
# 11. Find the average calories for each category and restaurant where the calories are greater than 500 and the protein is greater than 20 and the carbohydrates are less than 50 and the sodium is less than 1000 and the cholesterol is less than 100 and the sugars are less than 50 and the fiber is greater than 5 and the saturated fat is less than 10
query11 = 'SELECT Category, Restaurant, AVG(Calories) AS AvgCalories FROM menu WHERE Calories > 500 AND Protein > 20 AND Carbohydrates < 50 AND Sodium < 1000 AND Cholesterol < 100 AND Sugars < 50 AND Fiber > 5 AND SaturatedFat < 10 GROUP BY Category, Restaurant'

# display the results of the queries
pd.read_sql(query1, conn)


```
