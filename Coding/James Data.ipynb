```python
import numpy as np
import pandas as pd
import os
import sqlite3 as sql
```

```python
# Load the csv FastFoodNutritionMenuV2 from the resources folder. 

database_path = os.path.join('resources', 'FastFoodNutritionMenuV2.csv')

# display the first 5 rows of the data
data = pd.read_csv(database_path)
data.head()
```

```python
# import the data into a sqlite database
conn = sql.connect('fastfood.db')
data.to_sql('menu', conn, if_exists='replace', index=False)
# display the first 5 rows of the data from the sqlite database
query = 'SELECT * FROM menu LIMIT 5'
pd.read_sql(query, conn)
```
